1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
```

```python
def calculate_derivative(function):
  f_lin = function
  derivative_f_lin = sp.diff(f_lin, x)

  print(f"Похідна лінійної функції {function}: {derivative_f_lin}")
```

```python
calculate_derivative(10 * x - 5)
```

```python
x_space = np.linspace(0, 10, 100)
f = 10 * x_space - 5
df = np.full_like(x_space, 10)

# Створення графіку
plt.figure(figsize=(10, 6))
plt.plot(x_space, f, label='f(x) = 10x-5', color='blue')
plt.plot(x_space, df, label="f'(x) = 10", color='orange', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("Години навчання (x)")
plt.ylabel("Оцінка / Похідна")
plt.legend()
plt.grid(True)

plt.show()
```

Похідна f'(x)=10 означає, що кожна додаткова година навчання збільшує оцінку на 10 балів. Тобто, чим більше навчаєшся — тим вища оцінка.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
calculate_derivative(sp.sin(x))
```

```python
calculate_derivative(sp.ln(x))
```

```python
def plot_function_and_derivative(function, x_range, num_points=100):
    x = sp.Symbol('x')
    f = function
    df = sp.diff(f, x)

    # Числові функції
    f_func = sp.lambdify(x, f, 'numpy')
    df_func = sp.lambdify(x, df, 'numpy')

    # Область визначення
    x_vals = np.linspace(x_range[0], x_range[1], num_points)
    y_vals = f_func(x_vals)
    y_deriv_vals = df_func(x_vals)

    # Побудова графіку
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label=f'f(x) = {f}', color='red')
    plt.plot(x_vals, y_deriv_vals, label=f"f'(x) = {df}", color='orange', linestyle='--')
    plt.title("Графік функції та її похідної")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_function_and_derivative(sp.sin(x), x_range=(-np.pi, np.pi))
```

```python
plot_function_and_derivative(sp.sin(x), x_range=(-2*np.pi, 2*np.pi))
```

```python
plot_function_and_derivative(sp.ln(x), x_range=(0.1, 10))
```
